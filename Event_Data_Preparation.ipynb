```python
import pandas as pd

df = pd.read_csv('/content/Anonym_Daily_Events_22_23.csv')

df
```

```python
df['Start Event Type'].unique()
```

```python
df['Start Event Description'].unique()
```

```python
unique_combinations = df[['Start Event Type', 'Start Event Description']].drop_duplicates()
print(unique_combinations)
```

```python
df = df[~df['Start Event Type'].isin(['MTD', 'MTL'])]
df
```

```python
# Replace values in 'Start Event Type' column
df['Start Event Type'] = df['Start Event Type'].replace({'ELV': 'Full', 'IDV': 'Full', 'MDV': 'Empty', 'MLV': 'Empty'})

# Rename the column
df = df.rename(columns={'Start Event Type': 'Full / Empty'})

# Display the modified DataFrame
df
```

```python
# Replace values in 'Start Event Description' column
df['Start Event Description'] = df['Start Event Description'].replace({'Export Loaded on Vessel': 'Export',
                                                                     'Empty Loaded on vessel': 'Export',
                                                                     'Import Discharged from Vessel': 'Import',
                                                                     'Empty discharged from vessel': 'Import'})
df['Container Size'] =df['Equipment Type'].replace({'20DV': '20',
                                                       '40HC': '40',
                                                      '40DV': '40',
                                                      '20RE': '20',
                                                      '40HR':'40'})

df['Equipment Type'] = df['Equipment Type'].replace({'20DV': 'Dry',
                                                       '40HC': 'Dry',
                                                      '40DV': 'Dry',
                                                      '20RE': 'Reefer',
                                                      '40HR':'Reefer'})
# Rename the column
df = df.rename(columns={'Start Event Description': 'Export / Import'})

# Display the modified DataFrame
df
```

```python
df['Export / Import'].unique()
```

```python
# Convert 'Date' column to datetime if it's not already in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group data weekly by Date and create a 'weekyear' column
df['weekyear'] = df['Date'].dt.strftime('%G-%V')
df
```

```python
df
```

```python
# List of columns to exclude
exclude_cols = ['Unit Count', 'Date']

# Group by all columns except 'Unit Count' and 'Date' and aggregate
weekly_df = df.groupby([col for col in df.columns if col not in exclude_cols]).agg({
    'Unit Count': 'sum'
}).reset_index()

# Display the grouped and summarized DataFrame
weekly_df
```

```python
weekly_df.set_index('weekyear', inplace = True)
```

```python
weekly_df
```

```python
weekly_df = weekly_df[(weekly_df.index != '2021-52') & (weekly_df.index != '2023-52')]
```

```python
weekly_df
```

```python
weekly_df.reset_index()
```

```python
# Assuming your DataFrame is named df
pivoted_df = pd.pivot_table(weekly_df, values='Unit Count', index=['weekyear', 'Full / Empty', 'Equipment Type','Container Size', 'Start Event Location Full'],
                       columns=['Export / Import'], aggfunc="sum").reset_index()




# Displaying the pivoted DataFrame
pivoted_df
```

```python
# Check for NaN values in 'Export' and 'Import' columns
export_nulls = pivoted_df['Export'].isnull()
pivoted_df[export_nulls]

```

```python
import_nulls = pivoted_df['Import'].isnull()
pivoted_df[import_nulls]
```

```python
# Replace NaN values in 'Export' and 'Import' columns with 0
pivoted_df['Export'].fillna(0, inplace=True)
pivoted_df['Import'].fillna(0, inplace=True)

# Convert 'Export' and 'Import' columns to integer
pivoted_df['Export'] = pivoted_df['Export'].astype(int)
pivoted_df['Import'] = pivoted_df['Import'].astype(int)
```

```python
pivoted_df.rename(columns={'Export': 'Export Units', 'Import': 'Import Units', 'Start Event Location Full':'Location'}, inplace=True)
```

```python
pivoted_df
```

```python
pivoted_df['Imbalance'] = pivoted_df['Import Units'] - pivoted_df['Export Units']
```

```python
pivoted_df
```

```python
pivoted_df['Equipment Type'].value_counts()
```

```python
pivoted_df
```

```python
pivoted_df.to_csv("weekly_df.csv", index=False)

full_weekly_df = pivoted_df[pivoted_df['Full / Empty'] == 'Full']
full_weekly_df.drop(columns=['Full / Empty'], inplace=True)
full_weekly_df.to_csv("full_weekly_df.csv", index=False)


empty_weekly_df = pivoted_df[pivoted_df['Full / Empty'] == 'Empty']
empty_weekly_df.drop(columns=['Full / Empty'], inplace=True)
empty_weekly_df.to_csv("empty_weekly_df.csv", index=False)

```
